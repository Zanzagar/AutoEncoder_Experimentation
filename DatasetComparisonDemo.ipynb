# Dataset Generation and Visualization Comparison

This notebook demonstrates and compares both dataset formats supported by the autoencoder framework:

1. **Original Format**: Exact reproduction of `AutoEncoderJupyterTest.ipynb` dataset
2. **New Framework Format**: Enhanced dataset structure for future development

We'll generate both formats and visualize them to ensure consistency and verify that all functionality works correctly.

```python
# Setup and imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
import logging

# Set up logging and plotting
logging.basicConfig(level=logging.INFO)
plt.style.use('default')
sns.set_palette("husl")

# Set random seed for reproducibility
np.random.seed(42)

print("Dataset Comparison Demo")
print("=" * 50)
print("Testing both original and new framework dataset formats")
print("Random seed set to 42 for reproducibility")
```

```python
# Import the wrapper functions
from autoencoder_lib.data import generate_dataset, visualize_dataset

print("âœ… Successfully imported wrapper functions!")
print("\nAvailable dataset types:")
print("- 'layered_geological': Original format from AutoEncoderJupyterTest.ipynb")
print("- 'geological': New framework format with enhanced features")
```

## 1. Original Format: Layered Geological Dataset

This generates the exact same dataset as the original `AutoEncoderJupyterTest.ipynb` notebook with:
- Two classes: `consistent_layers` and `variable_layers`
- Individual PNG files saved to disk
- `.npy` metadata file with filenames and labels
- Compatible with all existing analysis code

```python
# Generate Original Format Dataset
print("ðŸ”„ Generating ORIGINAL format dataset...")
print("This creates the exact same format as AutoEncoderJupyterTest.ipynb")

original_dataset = generate_dataset(
    dataset_type="layered_geological",
    output_dir="comparison_original_dataset",
    num_samples_per_class=30,  # Moderate size for comparison
    image_size=64,
    random_seed=42,
    visualize=True,  # Show automatic visualizations
    force_regenerate=True
)

print(f"\nâœ… Original dataset generated successfully!")
print(f"ðŸ“ Dataset type: {type(original_dataset)}")
print(f"ðŸ”‘ Keys: {list(original_dataset.keys())}")
print(f"ðŸ“Š Total files: {len(original_dataset['filenames'])}")
print(f"ðŸ·ï¸  Classes: {original_dataset['label_names']}")
print(f"ðŸ“ Image size: {original_dataset['params']['image_size']}")

# Show file structure
print(f"\nðŸ“‚ File structure:")
sample_dir = os.path.dirname(original_dataset['filenames'][0])
print(f"   Directory: {sample_dir}")
print(f"   Sample files:")
for i, filename in enumerate(original_dataset['filenames'][:3]):
    print(f"     {os.path.basename(filename)}")
print(f"   Metadata: dataset_info.npy")
```

## 2. New Framework Format: Enhanced Geological Dataset

This generates the new framework format with:
- Five classes: `horizontal_layers`, `folded_layers`, `faulted_layers`, `intrusion_patterns`, `unconformity_patterns`
- Direct NumPy array access for better performance
- JSON metadata for human-readable configuration
- Extensible to additional geological pattern types

```python
# Generate New Framework Format Dataset
print("ðŸ”„ Generating NEW FRAMEWORK format dataset...")
print("This creates the enhanced geological dataset with 5 pattern types")

new_dataset = generate_dataset(
    dataset_type="geological",
    output_dir="comparison_new_dataset",
    num_samples_per_class=30,  # Same size for fair comparison
    image_size=64,
    random_seed=42,  # Same seed for consistency
    visualize=True,  # Show automatic visualizations
    num_layers_range=(3, 6),
    noise_level=0.05
)

print(f"\nâœ… New framework dataset generated successfully!")
print(f"ðŸ“ Dataset type: {type(new_dataset)}")
print(f"ðŸ”‘ Keys: {list(new_dataset.keys())}")
print(f"ðŸ“Š Total samples: {len(new_dataset['images'])}")
print(f"ðŸ·ï¸  Classes: {new_dataset['class_names']}")
print(f"ðŸ“ Image shape: {new_dataset['images'].shape}")

# Show metadata structure
print(f"\nðŸ“‹ Metadata structure:")
metadata = new_dataset['metadata']
print(f"   Generation params: {list(metadata['generation_params'].keys())}")
print(f"   Image size: {metadata['image_size']}")
print(f"   Random seed: {metadata['random_seed']}")
print(f"   Timestamp: {metadata['timestamp']}")
```

## 3. Detailed Visualization Analysis

Now let's perform comprehensive visualization analysis on both datasets to compare their characteristics and ensure consistency.

```python
# Analyze Original Format Dataset
print("ðŸ” Analyzing ORIGINAL format dataset...")

original_analysis = visualize_dataset(
    dataset_info=original_dataset,
    dataset_type="layered_geological",
    tsne_perplexity=20,  # Appropriate for dataset size
    tsne_random_state=42,
    show_statistics=True
)

print(f"\nðŸ“Š Original Dataset Analysis Results:")
print(f"   t-SNE projection shape: {original_analysis['tsne_projection'].shape}")
print(f"   PCA projection shape: {original_analysis['pca_projection'].shape}")
print(f"   Class distribution: {original_analysis['class_distribution']}")
print(f"   Classes analyzed: {original_dataset['label_names']}")
```

```python
# Analyze New Framework Format Dataset
print("ðŸ” Analyzing NEW FRAMEWORK format dataset...")

new_analysis = visualize_dataset(
    dataset_info=new_dataset,
    dataset_type="geological",
    tsne_perplexity=25,  # Slightly higher for more classes
    tsne_random_state=42,
    show_statistics=True
)

print(f"\nðŸ“Š New Framework Dataset Analysis Results:")
print(f"   t-SNE projection shape: {new_analysis['tsne_projection'].shape}")
print(f"   PCA projection shape: {new_analysis['pca_projection'].shape}")
print(f"   Class distribution: {new_analysis['class_distribution']}")
print(f"   Classes analyzed: {new_dataset['class_names']}")
```

## 4. Side-by-Side Comparison

Let's compare the key characteristics of both datasets to ensure they're consistent and working as expected.

```python
# Detailed Comparison
print("ðŸ“‹ COMPREHENSIVE DATASET COMPARISON")
print("=" * 60)

datasets = [
    ("ORIGINAL FORMAT", original_dataset, original_analysis),
    ("NEW FRAMEWORK FORMAT", new_dataset, new_analysis)
]

for name, dataset, analysis in datasets:
    print(f"\nðŸ”¸ {name}:")
    print(f"   ðŸ“ Data Structure: {type(dataset)}")
    
    if 'filenames' in dataset:  # Original format
        print(f"   ðŸ“Š Total Files: {len(dataset['filenames'])}")
        print(f"   ðŸ·ï¸  Classes: {dataset['label_names']} ({len(dataset['label_names'])} classes)")
        print(f"   ðŸ“ Image Size: {dataset['params']['image_size']}Ã—{dataset['params']['image_size']}")
        print(f"   ðŸ’¾ Storage: Individual PNG files + .npy metadata")
        
        # Load a sample image to get statistics
        from PIL import Image
        sample_img = np.array(Image.open(dataset['filenames'][0]))
        print(f"   ðŸ“ˆ Pixel Range: [{sample_img.min():.3f}, {sample_img.max():.3f}]")
        print(f"   ðŸ“Š Pixel Mean: {sample_img.mean():.4f}")
        print(f"   ðŸ“Š Pixel Std: {sample_img.std():.4f}")
        
    else:  # New format
        images = dataset['images']
        print(f"   ðŸ“Š Total Samples: {len(images)}")
        print(f"   ðŸ·ï¸  Classes: {dataset['class_names']} ({len(dataset['class_names'])} classes)")
        print(f"   ðŸ“ Image Shape: {images.shape}")
        print(f"   ðŸ’¾ Storage: NumPy arrays + JSON metadata")
        print(f"   ðŸ“ˆ Pixel Range: [{images.min():.3f}, {images.max():.3f}]")
        print(f"   ðŸ“Š Pixel Mean: {images.mean():.4f}")
        print(f"   ðŸ“Š Pixel Std: {images.std():.4f}")
    
    # Analysis results
    print(f"   ðŸ” t-SNE Shape: {analysis['tsne_projection'].shape}")
    print(f"   ðŸ” PCA Shape: {analysis['pca_projection'].shape}")
    print(f"   ðŸ“ˆ Class Distribution: {analysis['class_distribution']}")

print(f"\nâœ… Both datasets generated and analyzed successfully!")
print(f"ðŸŽ¯ Ready for autoencoder experimentation!")
```

## 5. Visual Sample Comparison

Let's display sample images from both datasets to visually verify the quality and consistency of the generated patterns.

```python
# Visual Sample Comparison
fig, axes = plt.subplots(2, 6, figsize=(18, 6))
fig.suptitle('Dataset Sample Comparison', fontsize=16, fontweight='bold')

# Original format samples
print("ðŸ“¸ Displaying sample images from both datasets...")

# Load original format samples
from PIL import Image
original_samples = []
original_labels = []
for i, (filename, label) in enumerate(zip(original_dataset['filenames'][:6], original_dataset['labels'][:6])):
    img = np.array(Image.open(filename))
    original_samples.append(img)
    original_labels.append(original_dataset['label_names'][label])

# Display original format samples
for i in range(6):
    axes[0, i].imshow(original_samples[i], cmap='viridis')
    axes[0, i].set_title(f"Original: {original_labels[i]}", fontsize=10)
    axes[0, i].axis('off')

# Display new format samples
new_samples = new_dataset['images'][:6]
new_labels = [new_dataset['class_names'][label] for label in new_dataset['labels'][:6]]

for i in range(6):
    axes[1, i].imshow(new_samples[i], cmap='viridis')
    axes[1, i].set_title(f"New: {new_labels[i]}", fontsize=10)
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()

print(f"âœ… Visual comparison complete!")
print(f"ðŸ” Both datasets show clear geological patterns")
print(f"ðŸ“Š Original format: 2 classes with layered variations")
print(f"ðŸ“Š New format: 5 classes with diverse geological features")
```

## 6. Loading Existing Datasets

Let's also test the ability to load and analyze existing datasets to ensure backward compatibility.

```python
# Test loading existing datasets
print("ðŸ”„ Testing dataset loading capabilities...")

# Check for existing datasets
existing_datasets = []
dataset_paths = [
    'layered_geologic_patterns_dataset/dataset_info.npy',
    'comparison_original_dataset/dataset_info.npy',
    'comparison_new_dataset/dataset_info.json'
]

for path in dataset_paths:
    if os.path.exists(path):
        existing_datasets.append(path)
        print(f"âœ… Found: {path}")

if existing_datasets:
    print(f"\nðŸ” Testing loading of existing dataset: {existing_datasets[0]}")
    
    # Determine dataset type based on file extension
    if existing_datasets[0].endswith('.npy'):
        dataset_type = "layered_geological"
    else:
        dataset_type = "geological"
    
    # Load and analyze
    loaded_analysis = visualize_dataset(
        dataset_path=existing_datasets[0],
        dataset_type=dataset_type,
        tsne_perplexity=20,
        tsne_random_state=42,
        show_statistics=True
    )
    
    print(f"âœ… Successfully loaded and analyzed existing dataset!")
    print(f"ðŸ“Š Loaded dataset analysis:")
    print(f"   t-SNE shape: {loaded_analysis['tsne_projection'].shape}")
    print(f"   Class distribution: {loaded_analysis['class_distribution']}")
    
else:
    print("â„¹ï¸  No existing datasets found - this is normal for a fresh setup")

print(f"\nðŸŽ¯ Dataset loading functionality verified!")
```
